```python
import json

def parse_json(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)

    answers = {}
    prompts = {}
    for sample in data["samples"]:
        question_id = sample["id"]
        correct_answer = sample["target"]
        model_answer = sample["scores"]["choice"]["answer"]
        # Сам промпт
        prompt = None
        for m in sample["messages"]:
            if m["role"] == "user":
                prompt = m["content"]

        answers[question_id] = {
            "correct": correct_answer,
            "model": model_answer,
            "is_correct": (model_answer == correct_answer)
        }
        prompts[question_id] = prompt

    return answers, prompts
```

```python
file1 = "/content/results/gpt-4o-mini/2025-07-20T05-22-57+00-00_arc-challenge_ERJpSbwha6TWtnvFaNueyU.json"
file2 = "/content/results/gpt4.1-nano/2025-07-20T05-25-00+00-00_arc-challenge_4UffAzbF9ssUXk5spN6amM.json"
```

```python
answers_1, prompts_1 = parse_json(file1)
answers_2, prompts_2 = parse_json(file2)


only_first_correct = []
only_second_correct = []

for qid in answers_1.keys():
    first = answers_1[qid]
    second = answers_2[qid]

    if first["is_correct"] and not second["is_correct"]:
        only_first_correct.append(qid)
    elif second["is_correct"] and not first["is_correct"]:
        only_second_correct.append(qid)


print("\n--- Questions answered correctly by ONLY the FIRST model ---")
for qid in only_first_correct:
    print(f"\nID: {qid}")
    print(prompts_1[qid])


print("\n--- Questions answered correctly by ONLY the SECOND model ---")
for qid in only_second_correct:
    print(f"\nID: {qid}")
    print(prompts_1[qid])
```

```python
def load_answers(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)

    answers = {}
    questions = {}
    choices = {}
    for sample in data["samples"]:
        question_id = sample["id"]
        correct_answer = sample["target"]
        model_answer = sample["scores"]["choice"]["answer"]
        question_text = sample["input"]
        answer_choices = sample.get("choices", [])

        answers[question_id] = {
            "correct": correct_answer,
            "model": model_answer,
        }
        questions[question_id] = question_text
        choices[question_id] = answer_choices
    return answers, questions, choices
```

```python
answers1, questions1, choices1 = load_answers(file1)
answers2, questions2, choices2 = load_answers(file2)

both_wrong = []

for qid in answers1:
    correct = answers1[qid]["correct"]
    answer1 = answers1[qid]["model"]
    answer2 = answers2[qid]["model"]

    if answer1 != correct and answer2 != correct:
        both_wrong.append(qid)
```

```python
print("\nQuestions where both models got it wrong:\n")
for qid in both_wrong:
  question_text = questions1[qid]
  correct = answers1[qid]["correct"]
  answer1 = answers1[qid]["model"]
  answer2 = answers2[qid]["model"]
  answer_choices = choices1.get(qid, [])

  print(f"ID: {qid}")
  print(f"Question:\n{question_text}\n")
  if answer_choices:
    print(f"Answer options: {', '.join(answer_choices)}\n")
  print(f"Correct answer: {correct}")
  print(f"Answer of the first model: {answer1}")
  print(f"Answer of the second model: {answer2}")
  print('-' * 80)
```
